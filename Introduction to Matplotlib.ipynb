# Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the python method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease [heart_disease["age"]> 50]
over_50.head()
```

```python
len(over_50)
```

```python
#Pyplot method
over_50.plot (kind = "scatter", x = 'age', y = 'chol', c = 'target');
plt.show();
```

```python
## OO method
fig, ax = plt. subplots(figsize = (10,6))
over_50.plot(kind= "scatter", x = "age", y = "chol", c = "target", ax = ax);
plt.show();

#ax.set_xlim([45,100]);
```

```python
over_50.target.values
```

```python
# OO method from scratch 
fig, ax = plt.subplots(figsize = (10, 6));

#Plot the data 
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"]); 

#Customize the plot 
ax.set (title = "Heart Disease and Cholesterol Levels",
xlabel = "Age", ylabel = "Cholesterol");

# Add a legend: scatter finds on target.values  what colors are
ax.legend (*scatter.legend_elements(), title = "Target");

# How to add a horizontal line in matplotlib plot
ax.axhline(over_50["chol"].mean(),linestyle = "--");
plt.show()
```

```python
#Subplot of chol, age, thalach 

fig,(ax0,ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10),sharex = True);

#Plot the data 
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"]); 

# Customisze ax0
ax0.set(title = "Heart Disease and Cholesterol Levels", ylabel = "Cholesterol");

# Add a legend to ax0
ax0.legend (*scatter.legend_elements(),title = "Target");

#Add a meanline
ax0.axhline (y= over_50["chol"].mean(), linestyle = "--");

#Add data to ax1
scatter = ax1.scatter (x = over_50["age"], y = over_50["thalach"], c = over_50["target"]);

#Customize ax1
ax1.set (title = "Heart Disease and Max Heart Rate", xlabel = "Age", ylabel = "Max Heart Rate");

# Add a legend to ax1
ax1.legend (*scatter.legend_elements(),title = "Target");

# Add a meanline
ax1.axhline(over_50["thalach"].mean(),linestyle = "--");

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = "bold");
```

```python
# Set the style 
plt.style.use('seaborn-v0_8-whitegrid');

# OO method from scratch 
fig, ax = plt.subplots(figsize = (10, 6));

#Plot the data 
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"],cmap ="winter"); #this changes the colour scheme

#Customize the plot 
ax.set (title = "Heart Disease and Cholesterol Levels",
xlabel = "Age", ylabel = "Cholesterol");

# Add a legend: scatter finds on target.values  what colors are
ax.legend (*scatter.legend_elements(), title = "Target");

# How to add a horizontal line in matplotlib plot
ax.axhline(over_50["chol"].mean(),linestyle = "--");
plt.show()
```

```python
#Customizing the y and x axis limitations 

#Subplot of chol, age, thalach 

fig,(ax0,ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10),sharex = True);

#Plot the data 
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"], cmap = "winter"); 

# Customisze ax0
ax0.set(title = "Heart Disease and Cholesterol Levels", ylabel = "Cholesterol");

#Change the x limits
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend (*scatter.legend_elements(),title = "Target");

#Add a meanline
ax0.axhline (y= over_50["chol"].mean(), linestyle = "--");

#Add data to ax1
scatter = ax1.scatter (x = over_50["age"], y = over_50["thalach"], c = over_50["target"], cmap = "winter");

#Customize ax1
ax1.set (title = "Heart Disease and Max Heart Rate", xlabel = "Age", ylabel = "Max Heart Rate");

ax1.set_xlim([50,80]);
ax1.set_ylim([60,200]);
# Add a legend to ax1
ax1.legend (*scatter.legend_elements(),title = "Target");

# Add a meanline
ax1.axhline(over_50["thalach"].mean(),linestyle = "--");

#Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = "bold");
```
